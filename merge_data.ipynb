```python
import pandas as pd
import numpy as np

df = pd.read_csv('Dataset.csv')
from_data = pd.read_csv('from.csv')
to_data = pd.read_csv('to.csv')

from_data['f0_'] = pd.to_datetime(from_data.f0_, format = "%Y-%m-%d")
to_data['f0_'] = pd.to_datetime(to_data.f0_, format = "%Y-%m-%d")

from_data['f0_'] = from_data['f0_'].apply(lambda x: x.strftime('%m/%d/%Y'))
to_data['f0_'] = to_data['f0_'].apply(lambda x: x.strftime('%m/%d/%Y'))

df['Date(UTC)'] = pd.to_datetime(df['Date(UTC)'], format = "%m/%d/%Y")
df['Date(UTC)'] = df['Date(UTC)'].apply(lambda x: x.strftime('%m/%d/%Y'))
to_data.head()
from_data.head()


print(from_data.columns)
from_unique_addresses = from_data['from_address'].unique()
to_unique_addresses = to_data['to_address'].unique()

for addr in from_unique_addresses:
    filter1 = from_data['from_address'] == addr
    temp = from_data.where(filter1)
    print(temp.dtypes)
    temp['fromSum_'+addr[-4:]] = temp['f1_'] * (10**(-18)) * temp['tx_count']
    #from_dataframes[addr] = temp
    df =  df.merge(temp, how='left', left_on='Date(UTC)',right_on = 'f0_').drop_duplicates()
    df.drop(['f0_', 'from_address', 'f1_'] ,axis=1,inplace= True)
    df.rename(columns={'tx_count':  'from_tx_count_' + addr[-4:]},inplace = True)

print(df)
```

```python
for addr in to_unique_addresses:
    filter1 = to_data['to_address'] == addr
    temp = to_data.where(filter1)
    temp = temp.astype({'f1_':'float64'})
#     temp['f1_'] = temp['f1_'].to_numeric()
    temp['toSum_'+addr[-4:]] = temp['f1_'] * (10**(-18)) * temp['tx_count']
    #from_dataframes[addr] = temp
    df =  df.merge(temp, how='left', left_on='Date(UTC)',right_on = 'f0_').drop_duplicates()
    df.drop(['f0_', 'to_address', 'f1_'] ,axis=1,inplace= True)
    df.rename(columns={'tx_count':  'to_tx_count_' + addr[-4:]},inplace = True)

df['Date(UTC)'] = pd.to_datetime(df['Date(UTC)'], format = "%m/%d/%Y")
df = df.sort_values(by = 'Date(UTC)', ascending = True)
print(df)
```

```python
df.to_csv('final_dataset.csv')
```
